```python
import pandas as pd
```

```python
df = pd.read_csv("covid.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

india = df[df['location'] == 'India']

plt.figure(figsize=(12,6))
plt.plot(india['date'], india['new_cases'], color='blue')
plt.title("Daily New COVID-19 Cases in India")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.show()
```

```python
countries = df[df['location'].isin(['India','United States'])]

plt.figure(figsize=(12,6))
for country in ['India','United States']:
    subset = countries[countries['location'] == country]
    plt.plot(subset['date'], subset['new_cases'], label=country)

plt.title("COVID-19 Cases: India vs USA")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Select numeric columns
numeric_df = df[['new_cases','new_deaths','new_vaccinations']].dropna()

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
